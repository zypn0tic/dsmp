`Problem-1:` Write a Python function that takes a list and returns a new list with unique elements of the first list.

`Problem-2:`Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
# Write code here
def sort_hyphen_sequence(seq):
    words = seq.split("-")           
    sorted_words = sorted(words)    
    return "-".join(sorted_words)    


input_seq = "green-red-yellow-black-white"
output_seq = sort_hyphen_sequence(input_seq)
print(output_seq) 
````

`Problem 3:` Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
# Write code here
sample = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'

def count_cases(sample): 
    count = 0 
    for char in sample: 
        if char.isupper(): 
            count += 1
    return count
count = count_cases(sample)
print('The number of upper cases is : ', count)
print('The number of lower cases is : ', len(sample) - count)
````

`Problem 5:` Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
# Write code here
a = int(input("Enter any number : "))
num = 1
fact = []

for num in range(1, a):
    if (a % num) == 0:
        fact.append(num)

print("The give number is perfect", a) if (sum(fact) == a) else print("The number is not perfect", a)    
````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
# Write code here
inp = 'hello how are you i am fine thank you'
word_lst = inp.split()
count = {}

for word in word_lst:
    if word in count: 
        count[word] += 1
    else:
        count[word] = 1

most_word = max(count, key=count.get)
print(most_word, "->", count[most_word])
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
def histogram(nums, bin_size=10):
    hist = {}

    for value in nums:
        bin_start = ((value - 1) // bin_size) * bin_size + 1
        bin_end = bin_start + bin_size - 1
        label = f"{bin_start}-{bin_end}"

        if label in hist:
            hist[label] += 1
        else:
            hist[label] = 1

    return hist


inp = [13, 42, 15, 37, 22, 39, 41, 50]
print(histogram(inp))
````

`Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
# Write code here 
def nearest_point(points, query):
    qx, qy = query

    min_distance = float('inf')
    nearest = None

    for x, y in points:
        distance_sq = (x - qx) ** 2 + (y - qy) ** 2

        if distance_sq < min_distance:
            min_distance = distance_sq
            nearest = (x, y)

    return nearest


# Example
points = [(1, 1), (2, 2), (3, 3), (4, 4)]
query = (0, 0)

result = nearest_point(points, query)
print(f"Nearest to {query} is {result}")
````

`Problem-10`:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

````python
# Write code here
def bag_of_words(sentences):
    bow = {}

    for sentence in sentences:
        words = sentence.split()   # split sentence into words

        for word in words:
            if word in bow:
                bow[word] += 1
            else:
                bow[word] = 1

    return bow


# Example usage
sentences = [
    "i love machine learning",
    "machine learning is fun",
    "i love python"
]

result = bag_of_words(sentences)
print(result)
````

`Problem-14`: Use reduce to convert a 2D list to 1D

````python
# Write code here
from functools import reduce
from operator import add

arr_2d = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Using a lambda function
flattened_arr_lambda = reduce(lambda x, y: x + y, arr_2d)

# Using operator.add (slightly more efficient)
flattened_arr_add = reduce(add, arr_2d)

print(f"Original array: {arr_2d}")
print(f"Flattened array (lambda): {flattened_arr_lambda}")
print(f"Flattened array (operator.add): {flattened_arr_add}")
````
